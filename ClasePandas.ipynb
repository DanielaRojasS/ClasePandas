```python
import pandas as pd

# cargar el archivo
url = '/content/Precio_maximo_de_venta_de_los_medicamentos_por_presentacion_comercial_20240419.csv'
df = pd.read_csv(url)
```

```python
# mostrar las primeras filas
print(df.head())
```

```python
# mostrar los ultimos datos del df
tail = df.tail()
print(tail)
```

```python
# calcular estadistica descriptiva
print(df.describe())
```

```python
# filtrar los datos
filtro = df['Descripción Principio Activo'] == 'Indapamida'
datos_filtrados = df[filtro]
print(datos_filtrados)
```

```python
# revisar cantidad de columnas y filas
shape = df.shape
print(shape)
```

```python
# valores unicos
unique = df['Descripción Principio Activo'].unique()
#df['Descripción Principio Activo'].drop_duplicates()
print(unique)
```
